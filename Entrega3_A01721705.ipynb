**Explicacion de algoritmos**

El primer algoritmo que encontramos es el de dar de alata a un empleado, este sirve para ingresar los datos nuevos de un empleado y se guardara en la lista de empleados, este te pide por el id, el nombre, la edad, el peso y la estatura del empleado, ademas te pregunta si el empleado tiene algunas enfermedades donde el la persona las indicara poniendo "1" si si la tiene y un "0" si no la tiene.

El segundo es el de calcular IMC del empleado, aqui se desplegara la informacion de la lista y esta mas enfocada en el peso y la altura, el programa encontrara el Indice de Masa Corporal del Empleado usando la formula de peso dividido por la altura al cuadrado, eso nos dara el IMC y se desplegara el numero y tambien si tiene un peso normal o si tiene algun tipo de delgadez o algun tipo de obesidad dependiendo del resultado del IMC, ademas se mostrara una tabla mostrando el porcentaje de empleados que tienen los diferentes tipos de pesos en particular.

El tercero es el de cambiar informaciom del empleado, cuando esta opcion es seleccionada, se preguntara por el ID del empleado que se quiere modificar, luego se mostrara la informacion del empleado y se preguntara si esta quiere ser cambiada, si es aceptado se preguntara cada uno de los datos y cuando ya sean modificado se cambuaran y esta nueva lista sera guardada en la lista de empleados.

El cuarto es la de consultar a un empleado, en esta opcion se preguntara por el ID del empleado que se desea consultar y si el ID es valido el programa mostrara los datos de ese empleado en particular, pero si el ID no es valido se le hara saber a la persona y se le indicara que ingrese un ID que sea valido o correcto.

El quinto es el de crear un reporte de empleados donde al seleccionar esta opcion se mostrara la lista completa de los empleados ya siendo actualizada y se podra ver toda la informacion que hay en la lista, tambien se mostrara una parte donde se puede ver cuantos empleados tienen lso diferentes tipos de enfermedades y el porcentaje de empleados que padecen de esa enfermedad.

El sexto es el de salida, esta opcion es para cuando la persona haya terminado pueda salir de el menu y este desplegara un mensaje diciendo "Muchas Gracias".


```python




lista_empleados = ["Teclea el id del empleado: ",
                   "Teclea el nombre del empleado: ",
                   "Teclea la edad del empleado: ",
                   "Teclea el peso del empleado (Kg): ",
                   "Teclea la altura del empleado (metros): ",
                   "1 - Si tiene diabetes, 0 - No tiene diabetes: ",
                   "1 - Si tiene hipertensión, 0 - No tiene hipertensión: ",
                   "1 - Si tiene enfermedad de corazón, 0 - No tiene enfermedad de corazón: ",
                   "1 - Si tiene cáncer, 0 - No tiene cáncer: ",
                   "Tabaquismo, 1 - Si, 0 - No: ",
                   "1 - Si tiene la vacuna, 0 - No tiene la vacuna: "]





global empleados
empleados = []

def leer_archivo(empleados):
  with open("empleados_guardados.txt", "r") as archivo:
    contenido = archivo.readlines()

    for elemento in contenido:
      linea = elemento.split(",")
      linea[10] = linea[10][:-1]
      empleados.append(linea)
      linea[3] = int(linea[3])
      linea[4] = float(linea[4])


def grabar_archivo(empleados):
  with open("empleados_guardados.txt", "w") as archivo:
    for empleado in empleados:
      empleado[1] = str(empleado[1])
      texto = ",".join(empleado[1])
      archivo.write(texto)


def alta_empleado():
  empleado = []
  
  info = 0
  while info < len(lista_empleados):

    info_nueva = input(lista_empleados[info])

    try:
      if info == 2 or info == 3:
        empleado.append(int(info_nueva))
      elif info == 4:
        empleado.append(float(info_nueva))
      else:
        empleado.append(info_nueva)

    except ValueError:
      print("Teclea el dato correcto")
      info = info - 1

    info += 1

  empleados.append(empleados)


def calcular_imc():
    
  

  contador_imc = [0, 0, 0, 0, 0, 0, 0, 0]
  tipo_imc = ["Delgadez severa", "Delgadez moderada", "Delgadez aceptable", "Normal", "Pre-obeso", "Obeso tipo I", "Obeso tipo II", "Obeso tipo III"]

  print(f'{"Clasificación IMC":<20}{"IMC":<10}{"ID":<10}{"Nombre":<15}{"Peso":<10}{"Altura":<10}{"Enfermedades":^77}{"Vacuna Covid":<12}')
  
  

  for empleado in empleados:
    
    
    id, nombre, edad, peso, altura, diabetes, hipertension, corazon, cancer, tabaquismo, vacuna_covid = empleado

    imc = peso / altura ** 2
    
    if imc < 16:
      clasificacion = tipo_imc[0]
      contador_imc[0] += 1
    elif 16 <= imc and imc < 17:
      clasificacion = tipo_imc[1]
      contador_imc[1] += 1
    elif 17 <= imc and imc < 18.50:
      clasificacion = tipo_imc[2]
      contador_imc[2] += 1
    elif 18.50 <= imc and imc < 25:
      clasificacion = tipo_imc[3]
      contador_imc[3] += 1
    elif 25 <= imc and imc < 30:
      clasificacion = tipo_imc[4]
      contador_imc[4] += 1
    elif 30 <= imc and imc < 35:
      clasificacion = tipo_imc[5]
      contador_imc[5] += 1
    elif 35 <= imc and imc < 40:
      clasificacion = tipo_imc[6]
      contador_imc[6] += 1
    else:
      clasificacion = tipo_imc[7]
      contador_imc[7] += 1

    imc = round(imc, 2)

    print(f'{clasificacion:<20}{imc:<10}{id:<10}{nombre:<15}{peso:<10}{altura:<10}{"Diabetes" if diabetes == "1" else "____________":15}{"Hipertensión" if hipertension == "1" else "____________":17}{"Corazón" if corazon == "1" else "____________":15}{"Cáncer" if cancer == "1" else "____________":15}{"Tabaquismo" if tabaquismo == "1" else "____________":15}{"Vacunado" if vacuna_covid == "1" else "____________":15}')
  
  print(f'\n{"Clasificación":27}{"Cantidad":10}{"Porcentaje":<10}')

  lista = []
  for num in range(len(contador_imc)):
    lista_temp = []
    lista_temp = [contador_imc[num], tipo_imc[num], round(contador_imc[num] / len(empleados) * 100)]
    lista.append(lista_temp)
  
  lista.sort(reverse = True)

  for info in lista:
    contador, etiqueta_imc, porcentaje = info
    print(f"{etiqueta_imc:20} {contador:7} {porcentaje:10}%")
  
  print(f'{"Total de empleados:":<10} {len(empleados):<10}')

  

def actualizar_empleado():

  

  nueva_id = input("Tecle el ID del empleado: ")

  for num in range(len(empleados)):

    empleado = empleados[num]

    id_empleado = empleado[0]

    if id_empleado == nueva_id:
      existe = True
      no_empleado = int(num)
      break
    else:
      existe = False

  if existe == True:
    print("El empleado existe")

    id, nombre, edad, peso, altura, diabetes, hipertension, corazon, cancer, tabaquismo, vacuna_covid = empleados[no_empleado]
    print(f'{"ID":<10}{"Nombre":<15}{"Peso":<10}{"Altura":<10}{"Enfermedades":^77}{"Vacuna Covid":<12}')
    print(f'{id:<10}{nombre:<15}{peso:<10}{altura:<10}{"Diabetes" if diabetes == "1" else "____________":15}{"Hipertensión" if hipertension == "1" else "____________":17}{"Corazón" if corazon == "1" else "____________":15}{"Cáncer" if cancer == "1" else "____________":15}{"Tabaquismo" if tabaquismo == "1" else "____________":15}{"Vacunado" if vacuna_covid == "1" else "____________":15}')
    conf = input("Si deseas cambiar los datos del empleado indica 'si'")
    if conf.lower() == "si":
      empleado = []

      info = 0
      while info < len(lista_empleados):

        info_nueva = input(lista_empleados[info])

        try:

          if info == 2 or info == 3:
            empleado.append(int(info_nueva))
          elif info == 4:
            empleado.append (float(info_nueva))
          else:
            empleado.append(info_nueva)
        
        except ValueError:
          print("Tecla el dato correcto: ")
          info = info - 1

        info += 1

      id, nombre, edad, peso, altura, diabetes, hipertension, corazon, cancer, tabaquismo, vacuna_covid = empleado

      print(f'La nueva información es :\n{"ID":<10}{"Nombre":<15}{"Peso":<10}{"Altura":<10}{"Enfermedades":^77}{"Vacuna Covid":<12}')
      print(f'{id:<10}{nombre:<15}{peso:<10}{altura:<10}{"Diabetes" if diabetes == "1" else "____________":15}{"Hipertensión" if hipertension == "1" else "____________":17}{"Corazón" if corazon == "1" else "____________":15}{"Cáncer" if cancer == "1" else "____________":15}{"Tabaquismo" if tabaquismo == "1" else "____________":15}{"Vacunado" if vacuna_covid == "1" else "____________":15}')

      conf = input("Si quieres cambiar la información teclea 'si ")
      if conf.lower() == "si":
        empleados[no_empleado] == empleado
  else: 
      print("El empleado no existe, asegurate que el ID sea valido")
      actualizar_empleado()
  

def consultar_empleado():

  

  while True:
    nueva_id = input("Teclea el ID del empleado que quieres consultar: ")

    for num in range(len(empleados)):

      empleado = empleados[num]

      id_empleado = empleado[0]

      if id_empleado == nueva_id:
        existe = True
        no_empleado = int(num)
        break
      else:
        existe = False

    if existe == True:
      print("El empleado existe")

      id, nombre, edad, peso, altura, diabetes, hipertension, corazon, cancer, tabaquismo, vacuna_covid = empleado

      print(f'{"ID":<10}{"Nombre":<15}{"Peso":<10}{"Altura":<10}{"Enfermedades":^77}{"Vacuna Covid":<12}')
      print(f'{id:<10}{nombre:<15}{peso:<10}{altura:<10}{"Diabetes" if diabetes == "1" else "____________":15}{"Hipertensión" if hipertension == "1" else "____________":17}{"Corazón" if corazon == "1" else "____________":15}{"Cáncer" if cancer == "1" else "____________":15}{"Tabaquismo" if tabaquismo == "1" else "____________":15}{"Vacunado" if vacuna_covid == "1" else "____________":15}')

    else:
      print("El empleado no existe, asegurate de que el ID sea valido")

      conf = input("Si quieres consultar otro empleado teclea 'si'")
      if conf.lower() == "si":
        continue
      else:
        break

def crear_reporte_empleado():

  
  contador_enfermedades = [0, 0, 0, 0, 0, 0]
  enfermedad = ["Diabetes", "Hipertensión", "Corazón", "Cáncer", "Tabaquismo", "Vacuna Covid"]

  print(f'{"ID":<10}{"Nombre":<15}{"Peso":<10}{"Altura":<10}{"Enfermedades":^77}{"Vacuna Covid":<12}')
  
  for empleado in empleados:

    id, nombre, edad, peso, altura, diabetes, hipertension, corazon, cancer, tabaquismo, vacuna_covid = empleado

    if diabetes == "1":
      contador_enfermedades[0] += 1
    if hipertension == "1":
      contador_enfermedades[1] += 1
    if corazon == "1":
      contador_enfermedades[2] += 1
    if cancer == "1":
      contador_enfermedades[3] += 1
    if tabaquismo == "1":
      contador_enfermedades[4] += 1
    if vacuna_covid == "1":
      contador_enfermedades[5] += 1
    
    print(f'{id:<10}{nombre:<15}{peso:<10}{altura:<10}{"Diabetes" if diabetes == "1" else "____________":15}{"Hipertensión" if hipertension == "1" else "____________":17}{"Corazón" if corazon == "1" else "____________":15}{"Cáncer" if cancer == "1" else "____________":15}{"Tabaquismo" if tabaquismo == "1" else "____________":15}{"Vacunado" if vacuna_covid == "1" else "____________":15}')
  
  print(f'\n{"Enfermedad":27}{"Cantidad":10}{"Porcentaje":<10}')

  lista = []

  for num in range(len(contador_enfermedades)):
    lista_temp = []
    lista_temp = [contador_enfermedades[num], enfermedad[num], round(contador_enfermedades[num] / len(empleados) * 100)]
    lista.append(lista_temp)

  lista.sort(reverse = True)

  for info in lista:
    contador, enfermedad, porcentaje = info
    print(f"{enfermedad:20} {contador:7} {porcentaje:10}%")

  print(f'{"Empleados totales:":<10} {len(empleados):<10}')

def menu():
  print("""Sistema de salud y bienestar del empleado:
    1.- Alta de empleado 
    2.- Calcular imc de empleado
    3.- Cambia información del empleado
    4.- Consulta un empleado
    5.- Reporte de todos los empleados
    6.- Salir""" )
  
  opcion = int(input("Teclea una de las siguientes opciones: "))

  return opcion

def salir():
  print("Muchas gracias")
  sigue = False

  return sigue
  
  

def main():

  leer_archivo(empleados)
  

  sigue = True

  while sigue:

    opcion = menu()

    if opcion < 1 or opcion > 6:
      print("Teclea un numero segun las opciones del menu")
      continue
    elif opcion == 1:
      alta_empleado()
    elif opcion == 2:
      calcular_imc()
    elif opcion == 3:
      actualizar_empleado()
    elif opcion == 4:
      consultar_empleado()
    elif opcion == 5:
      crear_reporte_empleado()
    else:
      sigue = salir()

main()
```
